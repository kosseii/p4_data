```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017

sous_nutrition['Année']= sous_nutrition['Année'].replace(['2012-2014'],2013)
sous_nutrition['Année']= sous_nutrition['Année'].replace(['2013-2015'],2014)
sous_nutrition['Année']= sous_nutrition['Année'].replace(['2014-2016'],2015)
sous_nutrition['Année']= sous_nutrition['Année'].replace(['2015-2017'],2016 )
sous_nutrition['Année']= sous_nutrition['Année'].replace(['2016-2018'],2017)
sous_nutrition['Année']= sous_nutrition['Année'].replace(['2017-2019'],2018)

join=pd.merge(sous_nutrition, population)
joint=join.loc[join['Année']==2017,:]


```

```python
#Affichage du dataset


joint


```

```python
#Calcul et affichage du nombre de personnes en état de sous nutrition
prets_auto = joint.loc[(joint['sous_nutrition']>0)]
prets_auto
print('Nous avons accordé', prets_auto.shape[0], 'pay sous nutrition et ',
prets_auto['sous_nutrition'].sum()/1e9, 'milliards de personnes en état de sous nutrition')

prets_auto['sous_nutrition'].sum()*100/joint['Population'].sum()

```

```python
#Combien mange en moyenne un être humain ? Source => 
besoins = 1700 # besoins journaliers moyen  en kilocalories
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
joint3=pd.merge(population, dispo_alimentaire, how='inner', on='Zone')
joint3=joint3.loc[joint3['Année']==2017,:]

```

```python
#Affichage du nouveau dataframe
joint3
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
joint3['dispo_kcal']=joint3['Disponibilité alimentaire (Kcal/personne/jour)']*joint3['Population']*365
joint3['dispo_kcal2']=joint3['Nourriture']+joint3['Semences']+joint3['Traitement']+joint3['Aliments pour animaux']+joint3['Autres Utilisations']+joint3['Pertes']
dis_mon=joint3['dispo_kcal'].sum()
print('La disponibilité mondiale est de ',dis_mon,' kcal.')
print('La disponibilité mondiale est de ',dis_mon/,' kcal.')


```

```python
#Calcul du nombre d'humains pouvant être nourris
print("nombre d'humains pouvant être nourris",(dis_mon/besoins)/1e9, 'milliards de personnes ')
print("nombre d'humains pouvant être nourris",((dis_mon/365)/besoins)/1e9, 'milliards de personnes ')
```

```python
#Calcul du nombre d'humains pouvant être nourris avec les végétaux

print('le nombre d humains pouvant être nourris avec les végétaux est:',(dis_veg/(besoins*365))/1e9,'milliards person' )
```
